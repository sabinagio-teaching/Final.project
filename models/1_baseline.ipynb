```python
# =========================================================
# 1. IMPORTS
# =========================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


# =========================================================
# 2. PREPARE DATA
# =========================================================
df.index = pd.to_datetime(df.index)

target_col = "DE_load_actual_entsoe_transparency"

df_model = df[[target_col]].dropna().copy()

# Add lag features
lags = [6, 48,24, 72]

for lag in lags:
    df_model[f"lag_{lag}"] = df_model[target_col].shift(lag)

df_model = df_model.dropna().copy()


# =========================================================
# 3. FEATURES + TARGET
# =========================================================
lag_cols = [f"lag_{lag}" for lag in lags]

X = df_model[lag_cols]
y = df_model[target_col]

lag_sets = {
    "small": [1, 24],
    "daily": [1, 24, 48, 72],
    "weekly": [1, 24, 168],
    "large": [1, 24, 48, 72, 168, 336]
}


# =========================================================
# 4. TIME-BASED SPLIT
# =========================================================
split_idx = int(len(df_model) * 0.8)

X_train = X.iloc[:split_idx].copy()
X_test = X.iloc[split_idx:].copy()

y_train = y.iloc[:split_idx].copy()
y_test = y.iloc[split_idx:].copy()

train_index = X_train.index
test_index = X_test.index

print("Shapes after split:")
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)


# =========================================================
# 5. TRAIN BASELINE MODEL
# =========================================================
baseline_model = LinearRegression()
baseline_model.fit(X_train, y_train)


# =========================================================
# 6. PREDICT
# =========================================================
y_pred = baseline_model.predict(X_test)


# =========================================================
# 7. EVALUATE
# =========================================================
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("\n=== Baseline Model Results ===")
print(f"MAE:  {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R²:   {r2:.4f}")


# =========================================================
# 8. PREDICTIONS DATAFRAME
# =========================================================
predictions_df = pd.DataFrame({
    "actual_values": y_test.to_numpy().flatten(),
    "predicted_values": y_pred.flatten()
}, index=test_index)

predictions_df["residuals"] = (
    predictions_df["actual_values"] - predictions_df["predicted_values"]
)

predictions_df["abs_error"] = np.abs(predictions_df["residuals"])

print("\nFirst predictions:")
print(predictions_df.head())


# =========================================================
# 9. PLOT: ACTUAL VS PREDICTED OVER TIME
# =========================================================
n_points = 200

plt.figure(figsize=(14, 5))
plt.plot(test_index[:n_points], y_test.iloc[:n_points], label="Actual Values")
plt.plot(test_index[:n_points], y_pred[:n_points], label="Baseline Predictions")

plt.title("Baseline Model: Actual vs Predicted Values Over Time")
plt.xlabel("Time")
plt.ylabel("Electricity Load")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# =========================================================
# 10. PLOT: SPECIFIC TIME PERIOD
# =========================================================
# You can change these dates
start_date = predictions_df.index.min().strftime("%Y-%m-%d")
end_date = (predictions_df.index.min() + pd.Timedelta(days=7)).strftime("%Y-%m-%d")

period_df = predictions_df.loc[start_date:end_date]

plt.figure(figsize=(14, 5))
plt.plot(period_df.index, period_df["actual_values"], label="Actual Values")
plt.plot(period_df.index, period_df["predicted_values"], label="Baseline Predictions")

plt.title(f"Baseline Model: Actual vs Predicted ({start_date} to {end_date})")
plt.xlabel("Time")
plt.ylabel("Electricity Load")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# =========================================================
# 11. PLOT: SCATTER ACTUAL VS PREDICTED
# =========================================================
plt.figure(figsize=(6, 6))
plt.scatter(y_test.to_numpy().flatten(), y_pred.flatten(), alpha=0.5, label="Predictions")

min_val = min(y_test.min(), y_pred.min())
max_val = max(y_test.max(), y_pred.max())

plt.plot([min_val, max_val], [min_val, max_val], linestyle="--", label="Perfect Fit")

plt.title("Baseline Model: Actual vs Predicted")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


# =========================================================
# 12. PLOT: RESIDUAL DISTRIBUTION
# =========================================================
plt.figure(figsize=(10, 5))
plt.hist(predictions_df["residuals"], bins=30, alpha=0.7)

plt.title("Baseline Model: Residual Distribution")
plt.xlabel("Prediction Error (Actual - Predicted)")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()


# =========================================================
# 13. PLOT: RESIDUALS VS ACTUAL VALUES
# =========================================================
plt.figure(figsize=(10, 5))
plt.scatter(y_test.to_numpy().flatten(), predictions_df["residuals"], alpha=0.5)

plt.axhline(0, linestyle="--")

plt.title("Baseline Model: Residuals vs Actual Values")
plt.xlabel("Actual Values")
plt.ylabel("Residuals")
plt.grid(True)
plt.tight_layout()
plt.show()


# =========================================================
# 14. PLOT: ABSOLUTE ERROR OVER TIME
# =========================================================
plt.figure(figsize=(14, 5))
plt.plot(predictions_df.index[:n_points], predictions_df["abs_error"].iloc[:n_points])

plt.title("Baseline Model: Absolute Error Over Time")
plt.xlabel("Time")
plt.ylabel("Absolute Error")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# =========================================================
# 15. TOP ERROR TIMESTAMPS
# =========================================================
largest_errors = predictions_df.nlargest(10, "abs_error")

print("\nTop 10 error timestamps:")
print(largest_errors)


# =========================================================
# 16. PLOT: LARGEST ERROR POINTS IN PERIOD
# =========================================================
plt.figure(figsize=(14, 5))
plt.plot(period_df.index, period_df["actual_values"], label="Actual Values")
plt.plot(period_df.index, period_df["predicted_values"], label="Baseline Predictions")

for error_time in largest_errors.index:
    if error_time in period_df.index:
        plt.axvline(error_time, linestyle="--", alpha=0.6)

plt.title(f"Baseline Model: Largest Error Points Highlighted ({start_date} to {end_date})")
plt.xlabel("Time")
plt.ylabel("Electricity Load")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# =========================================================
# 17. OPTIONAL: MODEL COEFFICIENTS
# =========================================================
coef_df = pd.DataFrame({
    "Feature": lag_cols,
    "Coefficient": baseline_model.coef_
})

print("\nBaseline model coefficients:")
print(coef_df)
```

*CONCLUSIONS*

Increasing the number of lag features does not necessarily improve model performance. While additional lags provide more historical information, they can introduce redundancy and noise. In this case, the model performed best when including lag 168, which captures the weekly seasonal pattern in electricity demand. This suggests that selecting meaningful lags is more important than simply increasing their number.
