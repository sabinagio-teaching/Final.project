```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Load data
df = pd.read_csv("/Users/jussaragaspar/Downloads/Final project/time_series_60min_singleindex_filtered.csv")
df.head(20)
```

```python
df["total_demand"] = (
    df["DE_load_actual_entsoe_transparency"] +
    df["FR_load_actual_entsoe_transparency"] +
    df["BE_load_actual_entsoe_transparency"]
)
```

```python
df["total_generation"] = (
    df["DE_wind_generation_actual"] +
    df["DE_solar_generation_actual"] +
    df["FR_solar_generation_actual"] +
    df["BE_solar_generation_actual"]
)
```

```python
df["grid_balance"] = df["total_generation"] - df["total_demand"]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(df.index[:200], df["grid_balance"][:200])
plt.axhline(0, color='red', linestyle='--')

plt.title("Grid Balance (Generation - Demand)")
plt.ylabel("MW")
plt.xticks(rotation=45)
plt.show()
```

```python
df["shortage"] = df["grid_balance"] < 0
```

```python
shortage_rate = df["shortage"].mean()
print("Shortage rate:", shortage_rate)
```

⭐️The grid balance model evaluates whether electricity supply
can meet demand. Negative values indicate shortages,
highlighting potential stress on the grid.
